```python
import yfinance as yf
import pandas as pd
import requests

#Had to use Claude to help download the data as Wiki blocked intitial search
#This data is used to get the tickers of the SNP500 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

html = requests.get(url, headers=headers).text
sp500 = pd.read_html(html)[0]

#Creates a list of the first 50 tickers
tickers = sp500['Symbol'].tolist()[:500]

#Downloads price data from 2020-2023 for each of the Symbols in the list tickers and stores it to the data frame data
#Using auto_adjust=True as in yfinance it automatically adjusts historical stock prices for splits and dividends. This ensures that the price data reflects the true performance of the stock over time.
data = yf.download(tickers, start='2020-01-01', end='2024-01-01', auto_adjust=True)['Close']

returned_tickers = data.columns.tolist()
failed_tickers = [t for t in tickers if t not in returned_tickers]
print(failed_tickers)
```
